#### RESISTENZA DI UN NUMERO

La resistenza di un numero si ottiene moltiplicando tutte le cifre che lo compongono.

Ad esempio, se si moltiplicano tutte le cifre del numero 882 si ottiene 128 (8*8*2). 
Ripetendo l'operazione per 128 si ottiene 16 (16).
Ripetendo l'operazione per 16 si ottiene 6 (1*6). 
6  di una sola cifra, quindi non  pi possibile ripetere l'operazione. 
Dato che siamo riusciti a ripetere l'operazione per 3 volte, diciamo che 882 ha una "resistenza" di 3.




###### 1) Scrivere la funzione getResistenza(numero), che restituisce la resistenza di un intero passato come argomento.

```python
def getResistenza(numero): 
   scala = 0 # per il livello della multiplication
   n=numero # passiamo alla variabile n il numero iniziale
   while n > 9: # a 10 parte il primo numero con pi di una chiffra, con scala >=1
     sn= str(n)  # affine di costrure una lista avendo come componenti le chiffre del numero oggetto di studio costruamo una string.
     ln=list(sn)
     n = 1 # inizialisiamo n a 1 il nostro n(numero iniziale). n=0 blocherebbe l'iterazione
     for sn in ln:
        j = int(sn)
        n = n * j
     scala = scala + 1
   return ("La resistenza di {}  {}".format(numero,scala))
```

```python
getResistenza(77)
```

FREQUENZA DI UN NUMERO

Nell'array di interi [2,4,1,5,6,3,4,2,7,4,5,4,5] il numero 2  presente 3 volte. Diciamo quindi che 2 ha una "frequenza" di 3.
Nello stesso array il numero pi frequente  4 (con una frequenza di 4). Il secondo numero pi frequente  5 (con una frequenza di 3).


1) Scrivere la funzione getFrequente(array) che restituisce il numero pi frequente in un array di interi passato come argomento.

```python
def getFrequente(array):
 
    ar=array
    ar.sort()# facciamo un ordinamento dell'array 
 
    maxar = 1; # la frequenza massimale.
    resa = ar[0]; # estrazione degli elementi 
    counta = 1 #indicio di frequenza che parte da uno per ogni prima occorenza.
    lar = len(ar)
    for i in range(1, lar):
        if (ar[i] == ar[i - 1]):
            counta += 1 #ogni volta che si ripete la presenza di numero viene incrementato l'indicio di frequenza.
             
        else :
            if (counta > maxar): # la frequenza massimal possibile  uguale alla lunghezza dell'array.
                maxar = counta
                resa = ar[i - 1]
             
            counta = 1
    
    if (counta > maxar):
     
        counta = maxar
        resa = ar[lar - 1]
     
    return resa
 
```

```python
ar=[2,4,1,5,6,3,4,2,7,4,5,4,5]
getFrequente(ar)
```

#####  2) Scrivere la funzione getFrequenti(array, k) che restituisce un array contenente i primi k numeri pi frequenti in un array di interi passato come argomento.

```python
def kfre(arr, k):
  
  lak=len(arr)# lunghessa dell'array che corrisponde al numero dei suoi elementi.
  count= {} # dictionario il complesso, "key:value" intero e frequenza
  freq=[[]for i in range(lak+1)] #list dov' andremo a stoccare le frequenze per potergli contare in un secondo momento
  for n in arr:
    count[n]= 1+ count.get(n,0) # per la prima volta che appare una variabile il suo valore nel dictionario  uno
    #per le prossime volte che appare si deve incrementare la sua value cumulata di uno. 
  for n,c in count.items():# n per la key e c per la value.
    freq[c].append(n) #
  res=[]
  print("The {} intergers with the highest frequencies are".format(k))
  for i in range(len(freq)-1,0,-1):
    for n in freq[i]:
      res.append(n)
      if len(res) == k:
        return (res)
```
